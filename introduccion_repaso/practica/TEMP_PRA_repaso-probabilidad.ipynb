```python
# importar 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Proba teorica de obtener cara
pc = 0.5

# Cantidad de experimentos a realizar
num_exp = 20

# simular el experimento de arrojar una moneda
def flip_coin(N, p=0.5):
    prob = [p, (1 - p)]
    return np.random.choice(['C','S'], size=N, p=prob)

# acumulada de experimentos
flips=flip_coin(num_exp, pc)

# contar la cantidad de caras
num_caras = np.sum(flips == 'C')

# Mostrar resultados
print("Experimentos:", " ".join(flips))
print(f"Numero de caras: {num_caras}")
print(f'P(C) = {num_caras/num_exp} (Numbero de caras/Total experimentos)')
```

```python
#Simulation probability of flipping heads
ph = 0.5

# Number of coin flips to simulate
num_flips = 200

# Simulate Coin flips
def flip_coin(N, p=0.5):
    prob = [p, (1 - p)]
    return np.random.choice(['C','S'], size=N, p=prob)

# Accumulate flips
flips=flip_coin(num_flips, ph)

# Count Heads
num_heads = np.sum(flips == 'C')

# Display results
#print("Flips:", " ".join(flips))
print(f"Numero de Caras: {num_heads}")
print(f'P(x = C) = {num_heads/num_flips} (Numero de Caras/Total de Experimentos)')
```

```python

# Simulated probability of flipping heads
#ph = 0.5

# Number of heads to compute
total = 50 ; size = 1

# Sampling array to plot coin flips
attempts = np.arange(size, total + size, size)

# Simulate coin flips
flips = flip_coin(total, ph)

# Accumulate head frequencies
results = [np.sum(flips[:idx] == 'C')/np.float(idx) for idx in attempts]

# Make plot
fig, ax = plt.subplots(figsize=(12, 8))

# Plot simulated flips
ax.plot(attempts, results , 'o-', alpha=0.75, label='Simulacion')

# Plot theoretical expectation
ax.axhline(ph, 0, 1, color='r', alpha=0.75, label='Teoria')

# Annotate plot
ax.legend(fontsize=16)
ax.set_title('Frecuencia P(x = C)',  fontsize=20)
ax.set_xlabel('Numero de experimentos',  fontsize=18)
ax.set_ylabel('P(x = C)', fontsize=18) 

# Compute y limits (how far above or below P(H) to display)
yp = abs(max(results) - ph)
yn = abs(min(results) - ph)

y_delta = max(yp, yn)

ax.set_ylim(ph - y_delta, ph + y_delta)

sns.despine(trim=True)
plt.show()
```
