# Calculando $\pi$ con el método de Sampling Acceptance-Rejection

```python
import numpy as np
from math import pi
import matplotlib.pyplot as plt
```

```python
it = 7

def calcular_pi(it):

    # Sampleo:
    x = np.random.rand(it)
    y = np.random.rand(it)

    # Ecuación de rechazo:
    results = x**2 + y**2 <= 1
    
    # Contar hits:
    hits = np.count_nonzero(results == True)
    
    # Proporción:
    area = hits / it
    
    # Despeje pi de ecuación de área:
    return area * 4, x, y, results
```

```python
pival, _, _,_ = calcular_pi(7)

pival
```

```python
it = 10000

x = range(1, it, 100)
pival = [calcular_pi(i)[0] for i in x]
```

```python
print(f'primeros: {pival[0:10]} \n\núltimos: {pival[-10:-1]}')
```

```python
plt.plot(x, pival)
plt.plot(x, [pi] * len(x))
plt.ylabel("$\pi$")
plt.xlabel("iteraciones")
plt.show()
```

```python
_, x, y, results = calcular_pi(1000)

x0 = x[results]
y0 = y[results]
x1 = x[~results]
y1 = y[~results]

plt.scatter(x0, y0, color='b')
plt.scatter(x1, y1, color='r')
plt.ylabel("$y$")
plt.xlabel("$x$")
plt.show()
```

```python
print(f'Pi con {it} iteraciones: {pival[-1]}')
```
