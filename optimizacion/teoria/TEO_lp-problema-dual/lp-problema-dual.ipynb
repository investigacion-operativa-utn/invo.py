## Max en primal / Min en dual

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the projection function
def projection_function(x, y):
    return 307 * x + 550 * y

plt.subplots(figsize=(15, 10))

# Generate data for X and Y
x = np.linspace(0, 2.5, 1000)
y = np.linspace(0, 2.5, 1000)
X, Y = np.meshgrid(x, y)

# Calculate the projected values for each point
Z = projection_function(X, Y)

# Create a contour plot for the projection
plt.contourf(X, Y, Z, cmap='viridis', levels=range(0, 2200, 50))

# Add colorbar
plt.colorbar()

# Add constraint lines
constraint1 = (30 * x + 72 * y) >= 72
constraint2 = (23 * x + 21 * y) >= 33

plt.plot(x, (72 - 30 * x) / 72, label=r'$30x + 72y \leq 72$', color='white', linestyle ='dotted')
plt.plot(x, (33 - 23 * x) / 21, label=r'$23x + 21y \leq 33$', color='white')

# Fill the feasible region
plt.fill_between(x, np.maximum((72 - 30 * x) / 72, (33 - 23 * x) / 21), 2.5, color='white', alpha=0.3)

# Set labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Projection of z = 307x + 550y onto X-Y Plane with Constraints')

plt.xlim([0, 2.5])
plt.ylim([0, 2.5])

# Add legend
plt.legend()

# Show the plot
plt.show()
```
