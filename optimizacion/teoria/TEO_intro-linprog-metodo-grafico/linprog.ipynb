```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.colors import Normalize

# Crear datos para los ejes x e y
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)

# Calcular los valores correspondientes de z según la ecuación z = 72x + 33y
z = 72 * x + 33 * y

# Crear la figura y el eje 3D
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(111, projection='3d')

# Graficar el plano en 3D con la barra de colores
surf = ax.plot_surface(x, y, z, cmap='viridis', rstride=1, cstride=1,
                       linewidth=0, antialiased=False)

# Agregar la barra de colores
mappable = cm.ScalarMappable(norm=Normalize(vmin=z.min(), vmax=z.max()), cmap='viridis')
mappable.set_array([])
cbar = plt.colorbar(mappable, ax=ax, pad=0.1)
cbar.set_label('Valor de z')

# Etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Mostrar el gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear datos para la región factible
x = np.linspace(0, 25, 400)
y = np.linspace(0, 25, 400)
x, y = np.meshgrid(x, y)
eq1 = 30 * x + 23 * y
eq2 = 72 * x + 21 * y

# Crear la figura y el gráfico
fig, ax = plt.subplots(figsize=(12, 12))

# Graficar las restricciones 30x + 23y <= 307 y 72x + 21y <= 550
ax.contourf(x, y, eq1, levels=[-np.inf, 307], colors='blue', alpha=0.3)
ax.contourf(x, y, eq2, levels=[-np.inf, 550], colors='red', alpha=0.3)

# Etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')



plt.xlim([0, 15])
plt.ylim([0, 15])

# Mostrar la gráfica
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the projection function
def projection_function(x, y):
    return 72 * x + 33 * y

plt.subplots(figsize=(15, 10))

# Generate data for X and Y
x = np.linspace(0, 20, 1000)
y = np.linspace(0, 20, 1000)
X, Y = np.meshgrid(x, y)

# Calculate the projected values for each point
Z = projection_function(X, Y)

# Create a contour plot for the projection
plt.contourf(X, Y, Z, cmap='viridis', levels=range(0, 3000, 50))

# Add colorbar
plt.colorbar()

# Add constraint lines
constraint1 = (30 * x + 23 * y) <= 307
constraint2 = (72 * x + 21 * y) <= 550

plt.plot(x, (307 - 30 * x) / 23, label=r'$30x + 23y \leq 307$', color='white', linestyle ='dotted')
plt.plot(x, (550 - 72 * x) / 21, label=r'$72x + 21y \leq 550$', color='white')

# Fill the feasible region
plt.fill_between(x, np.minimum((307 - 30 * x) / 23, (550 - 72 * x) / 21), color='white', alpha=0.3)

# Set labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Projection of z = 72x + 33y onto X-Y Plane with Constraints')

plt.xlim([0, 20])
plt.ylim([0, 20])

# Add legend
plt.legend()

# Show the plot
plt.show()
```
